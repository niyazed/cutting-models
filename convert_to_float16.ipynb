## Convert all the weights Float32 to Float16

````python
# Converts float32 datatype to float16 
# per layer weights of given model

def convert_to_float16(weights):
    converted = model[weights].to(dtype = torch.float16)
    print(converted.dtype)
    return converted
````

````python
# initialize a ordered dictionary

float16_model = OrderedDict()
````

````python
for layer_name, weights in model.items():
    float16_model[layer_name] = convert_to_float16(layer_name)
#     print(layer_name)
````

````python
torch.save(float16_model, './float16_model.pth')
````
